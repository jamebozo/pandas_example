```python
# show mean company value's 
comp_grp['Market Value'].mean()
```

```python

comp_grp['Market Value'].agg(['median','mean']) 
```

```python
max_value = comp_grp['Market Value'].apply(lambda x: max(x))
max_value
#comp_grp['Industry Format'].value_counts().head(20)
```

```python
value_larger_than_100 = comp_grp['Market Value'].apply(lambda x: x > 100)
value_larger_than_100
```

```python
market_value_pd = pd.concat([max_value, value_larger_than_100], axis='columns',sort=False)
market_value_pd.columns=['value','>100']
market_value_pd
```
