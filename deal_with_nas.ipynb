```python
dic={'name':['james', np.nan, 'Nan', 'Chris', np.nan, 'mike', None], 
     'phone':['0910817850', np.nan,'091231234', np.nan,'091231222', '09888888888','missing'], 
     'email':['james@alcorlink.com', np.nan, 'tina@tclink.com', 'mike@alcorlink.com',None, np.nan, 'donald@alcorlink.com'],
    'age':['12','23','25','53','43','11','24']}
dic
```

```python
df = pd.DataFrame(dic)
df
```

```python
# clean None and NaN
df.dropna()
```

```python
# default settings of dropna
df.dropna(axis='index', how='any')
```

```python
# drop only if all values are NaN
df.dropna(axis='index', how='all')  # drop 1 row
```

```python
# if telephone is not important, nan is ok. But we must have email
df.dropna(axis='index', how='all',subset=['email'])
# now how only checks 'email' col
```

```python
# if telephone is not important, nan is ok. But we must have email
df.dropna(axis='index', how='all',subset=['email', 'phone'])
# now how only checks 'email' col
```

```python
# replace "custom Nan"
df.replace('missing',np.nan, inplace=True)
```

```python
# check where is na
df.isna()
```

```python
# replace to 0,(dealing with numbers)
df.fillna('MISSING')
```

```python
# show types of columns
df.dtypes
```

```python
df['age'] = df['age'].astype(int)
```
